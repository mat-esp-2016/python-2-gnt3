# Bibliotecas:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches
import glob
```

# Repetição:

```python
nome_arquivos = glob.glob("dados/*W-TAVG-Trend.txt")

#Começamos a repetição para todos os arquivos da variavel "nome_arquivos"
for arquivo in nome_arquivos:
    
    #Primeiro de tudo, criamos as variáveis 
    dados = np.loadtxt (arquivo, comments="%")
    anos = dados[:, 0]
    meses = dados[:, 1]
    ano_decimal = (meses - 1)/12 + anos
    anomalia = dados[:, 4]
    anomalia_10anos = dados [:,8]
    unc_10anos = dados [:, 9]
    unc_95pos = (anomalia_10anos + unc_10anos)
    unc_95neg = (anomalia_10anos - unc_10anos)
    
    #Agora começa a criação do gráfico
    plt.figure()
    
    plt.fill_between(anos, unc_95pos, unc_95neg, color = "#C0C0C0")
    plt.plot (ano_decimal, anomalia, c = "#000000")
    plt.plot (anos, anomalia_10anos, c = "red")
    
    #Definimos os nomes de cada eixo
    plt.xlabel ("Anos")
    plt.ylabel ("Temperatura °C")
    plt.title("Brasil")

    #Finalmente salvamos a figura em formato png na pasta gráficos com o nome dos próprios arquivos que extraimos os dados
    plt.savefig("graficos/" + arquivo[7:-3] + "png", format = "png")
```
