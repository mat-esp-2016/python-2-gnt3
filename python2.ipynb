# Bibliotecas:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches
import glob
```

# Repetição:

```python
nome_arquivos = glob.glob("dados/*W-TAVG-Trend.txt")

for arquivo in nome_arquivos:
    dados = np.loadtxt (arquivo, comments="%")
    anos = dados[:, 0]
    meses = dados[:, 1]
    ano_decimal = (meses - 1)/12 + anos
    anomalia = dados[:, 4]
    
    plt.figure()
    
    plt.plot (ano_decimal, anomalia, c = "#000000")
    
    plt.xlabel ("Anos")
    plt.ylabel ("Temperatura °C")
    plt.title("Brasil")


    plt.savefig("graficos/" + arquivo[7:-3] + "png", format = "png")
```
